```python
Abstract

Este proyecto utiliza el dataset “Melbourne Housing Snapshot” proveniente de Kaggle, que contiene información de más de 30.000 propiedades vendidas en Melbourne, Australia. La base de datos incluye características de las propiedades como tipo de vivienda, precio de venta, cantidad de dormitorios y baños, superficie de construcción y del terreno, ubicación (suburbio, distancia al centro), año de construcción, entre otras variables relacionadas con el entorno y la propiedad.

El objetivo del proyecto es explorar el mercado inmobiliario de Melbourne desde una perspectiva descriptiva y visual, respondiendo preguntas sobre la distribución de precios, la relación entre características de las propiedades y su valor, y la variación por suburbios y tipos de vivienda. Para ello se aplican técnicas de análisis univariado, bivariado y multivariado, utilizando herramientas como Matplotlib, Seaborn y pandas.
```
